## 探空的绘制

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
from skewt import SkewT
help(SkewT.Sounding)
```

```python
s = SkewT.Sounding('./data/58238_2014071112_uwyo.txt')
```

```python
help(s)
```

```python
s.plot_skewt(color='r')
plt.show()
```

```python
s.make_skewt_axes(tmin=-30.,tmax=40.,pmin=100.,pmax=1050.)
s.add_profile(color='b',bloc=0)
parcel=s.get_parcel()
s.lift_parcel(*parcel)
plt.show()
```

```python
S=SkewT.Sounding('./data/58238_2014071112_uwyo.txt')
T=SkewT.Sounding('./data/58203_2017050200_uwyo.txt')
S.make_skewt_axes()
S.add_profile(color='r',bloc=0)
S.soundingdata=T.soundingdata      # replace the sounding data in S with that from T
S.add_profile(color='b',bloc=0.5)
plt.show()
```
